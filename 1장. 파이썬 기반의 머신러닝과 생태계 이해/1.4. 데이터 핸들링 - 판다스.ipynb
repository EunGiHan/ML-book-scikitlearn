## 1.4.2. DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호변환

```python
import numpy as np

### 1차원 데이터를 DataFrame로 변환 : column이 1개만 필요
col_name1 = ['col1']
list1 = [1, 2, 3]
array1 = np.array(list1) #리스트를 ndarray로 변환
print('array1 shape : ', array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

# ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

```python
### 2차원 데이터를 DataFrame로 변환
col_name2 = ['col1', 'col2', 'col3']
list2 = [[1, 2, 3], [11, 12, 13]]
array2 = np.array(list2)
print('array2 shape : ', array2.shape)

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)

df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

```python
### 딕셔너리를 DataFrame로 변환 : 딕셔너리의 key를 칼럼명(문자열)으로, value를 칼럼 데이터(리스트, ndarray)로 변환
dict = {'col1':[1, 11], 'col2':[2, 12], 'col3':[3, 13]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

```python
### DataFrame을 ndarray로 변환 : values 이용
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shpae:', array3.shape)
print(array3)
```

```python
### DataFrame을 리스트로 변환 : tolist() 이용
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)
```

```python
### DataFrame을 딕셔너리로 변환 : to_dict() 이용, 'list'를 인자로 전달해 리스트형으로 반환
dict3 = df_dict.to_dict('list')
print('df_dict.values.to_dict() 타입:', type(dict3))
print(dict3)
```

## 데이터 셀렉션 및 필터링
* 넘파이의 []연산자와 유사한 부분도 있고 다른 부분도 있음에 주의

## apply lambda 식으로 데이터 가공
* apply 함수에 lambda 식을 결합해 DataFrame/Series의 레코드 별로 데이터를 가공하는 기능 제공
* 칼럼에 일괄적으로 데이터 가공하는 것이 속도 면에서 빠르거나/복잡한 데이터 가공 필요할 시 apply lambda 이용

```python
### lambda 함수 : lambda 식 이용 시 여러 개의 값을 입력 인자로 사용해야 할 때는 map() 함수 결합
a=[1, 2, 3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
### 판다스 DataFrame의 lambda 식
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x: len(x)) #Name 칼럼의 문자열 개수를 별도 칼럼으로 생성
titanic_df[['Name', 'Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
# lambda는 else if 지원하지 않으므로 중첩해 사용함
titanic_df['Age_cat'].value_counts()
```

```python
def get_category(age):
    cat = ''
    if age<=5 : cat = 'Baby'
    elif age<=12 : cat = 'Child'
    elif age<=18 : cat = 'Teenager'
    elif age<=25 : cat = 'Student'
    elif age<=35 : cat = 'Young Adult'
    elif age<=60 : cat = 'Adult'
    else : cat = 'Elderly'
    return cat

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x)) #함수를 lambda 식 안에 적용 가능
titanic_df[['Age', 'Age_cat']].head()
```
