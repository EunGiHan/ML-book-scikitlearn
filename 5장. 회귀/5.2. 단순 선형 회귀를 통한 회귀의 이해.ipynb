## 단순 선형 회귀
* 독립변수도, 종속변수도 각각 하나인 선형 회귀
* 예측값 Y-hat = w0 + w1X 이라면 회귀 계수는 기울기(w1), 절편(w0)

## 오류
* 잔차 : 실제값과 회귀 모델 차이에 따른 오류값
* 최적의 회귀 모델 = 전체 데이터의 잔차 합이 최소가 되는 모델 & 오류값 합이 최소가 되는 회귀 계수 찾기
* 오류 합의 계산 방식
  * 절댓값을 취하기 : Mean Absolute Error
  * 오류값의 제곱을 구해서 더하기 : RSS, Residual Sum of Square
* 일반적으론 미분 계산 등을 편하게 하려 Error^2 = RSS로 사용
* RSS를 최소로 하는 회귀 계수를 학습 통해 찾는 것이 머신러닝 핵심
* w변수인 회귀 계수가 중심 변수임. 독립변수 X, 종속변수 Y는 RSS에서 상수로 간주
* RSS는 학습 데이터 건수로 나누어 정규화된 식을 표현됨


## 비용함수
* 회귀에서 RSS는 비용(cost)임
* 비용함수(손실함수. loss function) : w변수인 회귀 계수로 구성되는 RSS
* 데이터를 계속 학습하며 비용함수가 반환하는 값인 오류값을 지속해서 감소시키고, 최종적으로 더이상 감소하지 않는 최소의 오류값 구함
