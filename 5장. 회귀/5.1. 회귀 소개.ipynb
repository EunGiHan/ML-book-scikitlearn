## 회귀(Regression)
* 회귀 분석
  * 데이터값이 평균과 같은 일정한 값으로 돌아가려는 경향을 이용한 통계학 기법
  * 여러 개의 독립변수와 한 개의 종속 변수 간의 상관관계를 모델링하는 기법
  * 선형회귀 식 : Y = W1X1 + W2X2 + ... + WnXn이라면 Y는 종속변수, X는 독립변수, W는 회귀계수(regression coefficients)
  * 독립변수는 피처, 종속변수는 결정값
  * 주어진 피처와 결정값 데이터 기반에서 학습을 통해 최적의 회귀계수를 찾아내는 것
* 지도학습에서의 회귀
  * 지도학습은 분류와 회귀 두 유형으로 나뉨
  * 분류는 예측값이 이산형 클래스값(카테고리 등), 회귀는 연속형 숫자값

## 회귀의 유형
* 회귀 계수의 선형, 비선형 // 독립변수의 개수 // 종속변수의 개수 따라 나눌 수 있음
* 회귀 계수가 선형인가? **선형 회귀** 비선형인가? **비선형 회귀**
* 독립 변수가 한 개인가? **단일 회귀** 여러 개인가? **다중 회귀**

## 선형 회귀
* 여러 회귀 중 가장 많이 사용됨
* 실제값과 예측값의 차이(오류 제곱값)을 최소화하는 직선형 회귀선을 최적화하는 방식
* 규제(regularization) : 일반적인 선형 회귀의 과적합 문제를 해결하기 위해 회귀 계수에 페널티 값을 적용하는 것
  * L1 규제 : 예측 영향력이 작은 피처의 회귀 계수를 0으로 만듦 -> 회귀 예측 시 피처가 선택되지 않게 함. 피처선택 기능
  * L2 규제 : 상대적으로 큰 회귀 계수 값의 예측 영향도를 감소시키기 위해 회귀 계수값을 더 작게 만드는 규제 모델
* 규제 방법에 따라 유형 나뉨
  * 일반 선형 회귀 : 예측값과 실제값의 RSS(residual sum of squares)을 최소화하는 회귀계수로 최적화함. 규제 적용 안함
  * 릿지(Ridge) : 선형회귀에 L2규제 추가
  * 라쏘(Lasso) : 선형회귀에 L1규제 추가
  * 엘리스틱넷(ElasticNet) : L2(게수 값 크기 조정), L1(피처 개수 줄임) 규제 함께 결함. 피처 많은 데이터셋에 적용
  * 로지스틱 회귀(Logistic Regression) : 분류에 사용되는 선형 모델
