# 협업 필터링
* 사용자가 아이템에 매긴 평점 정보, 상굼 구매 이력 등 사용자 행동 양식(user bahavior) 만을 기반으로 추천을 수행하는 것
* 목표 : 사용자-아이템 평점 매트릭스와 같은 축적된 사용자 행동 데이터를 기반 -> 사용자가 아직 평가하지 않은 아이템을 예측 평가(Predict Rating)하는 것
* 평가한 다른 아이템을 기반으로 평가하지 않은 아이템의 예측 평가를 도축하는 방식
* 최근접이웃, 잠재요인 방식 둘 다 사용자-아이템 평점 행렬만을 사용
* 사용자-아이템 평점 행렬
  * 행 = 개별 사용자, 열=개별 아이템
  * 사용자 아이디 행, 아이템 아이디 열 위치 = 해당하는 값이 평점 나타냄
  * 레코드 레벨 형태라면 판다스의 pivot_table() 함수 이용해 변경해야 함
  * 많은 아이템을 가지는 다차원 행렬
  * 희소행렬(sparse matrix) : 사용자가 아이템에 대한 평점을 매기는 경우가 많지 않아서

# 최근접 이웃 협업 필터링
* =메모리(memory) 협업 필터링
* 분류
  * 사용자 기반(User-User) : 당신과 비슷한 고객들이 다음 상품도 구했습니다.
    * 특정 사용자와 유사한 다른 사용자를 Top-N으로 선정해 그들이 좋아하는 아이템 추천
    * 특정 사용자와 타 사용자간 유사도(similarity) 측정한 뒤, 가장 유사도 높은 top-n 사용자 추출해 그들이 선호하는 아이템 추천
  * 아이템 기반(Item-Item) : 이 상품을 선택한 다른 고객들은 다음 상품도 구했습니다.
    * 아이템이 가진 속성과는 상관 없음
    * 사용자들이 그 아이템을 좋아하는지/싫어하는지의 평가 척도가 유사한 아이템을 추천하는 기준이 되는 알고리즘
* 사용자 기반 <=> 아이템 기반의 행/열이 서로 반대. 사용자 기반은 행이 사용자 열이 아이템, 아이템 기반은 행이 아이템, 열이 사용자
* 일반적으로 아이템 기반이 더 정확도 높음 -> 최근접 이웃 협업 필터링은 대부분 아이템 기반
* 코사인 유사도 : 추천 시스템의 유사도 측정에 가장 많이 적용됨(다차원 희소 행렬)
