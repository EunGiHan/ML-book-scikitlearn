# 추천 시스템의 유형
* 콘텐츠 기반 필터링(Content Based Filtering) : 사용자가 특정한 아이템 매우 선호할 때, 그 아이템과 비슷한 콘텐츠 가진 다른 아이템을 추천하는 방식
* 협업 필터링(Collaborate Filtering)
  * 최근접 이웃(Nearest Neighbor) 협업 필터링
  * 잠재 요인(Latent Factor) 협업 필터링
* 초창기에는 콘텐츠 기반, 최근접이웃협업이 주로 사용 -> 넷플릭스 추천 시스템 경연 대회에서 행렬분해 기법 이용한 잠재요인협업 우승하며 이를 기반으로 추천 시스템 구축
* 개인화 특성 좀 더 강화시키려 하이브리드 방식으로 콘텐츠+협업기반 결합해 사용하기도
