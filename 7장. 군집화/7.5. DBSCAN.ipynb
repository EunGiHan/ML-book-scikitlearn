# DBSCAN(Density Based Spatial Clustering of Applications with Noise)
* 특징
  * 밀도 기반 군집화의 대표적 알고리즘
  * 간단하고 직관적임
  * 데이터 분포가 기하하적으로 복잡한 데이터셋에도 효과적
* 중요 파라미터
  * 입실론 주변 영역(epsilon) : 개별 데이터를 중심으로 입실론 반경을 가지는 원형의 영역
  * 최소 데이터 개수(min points) : 개별 데이터의 입실론 주변 영역에 포함되는 타 데이터의 개수
* 데이터 포인트 : 입실론 주변 영역 내에 포함되는 최소 데이터의 개수를 충족시키는가 아닌가에 따라 분류(책 442-443그림 및 설명)
  * 핵심 포인트(core point) : 주변 영역 내에 최소 데이터 이상의 타 데이터 있는 데이터
  * 이웃 포인트(neighbor point) : 주변 영역 내에 위치한 타 데이터
  * 경계 포인트(border point) : 주변 영역 내에 최소 데이터 개수 이상의 이웃 포인트는 없음 & 핵심 포인트를 이웃 포인트로 가지고 있는 데이터
  * 잡음 포인트(noise point) : 최소 데이터 개수 이상의 이웃 포인트 없음 & 핵심포인트도 이웃 포인트로 없는 데이터
* 입실론 주변 영역의 최소 데이터 개수를 포함하는 밀도 기준을 충족시키는 데이터인 핵심 포인트를 연결하면서 군집화를 구성하는 방식
* 사이킷런 DBSCAN 클래스의 주요 초기화 파라미터
  * eps : 입실론 주변 영역의 반경
  * min_smaples : 핵심 포인트가 되려 입실론 주변 영역 내에 포함되어야 할 데이터의 최소 개수. 자신 데이터를 포함

# DBSCAN 적용하기 - 붓꽃 데이터 세트

```python
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

iris = load_iris()
feature_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
irisDF = pd.DataFrame(data=iris.data, columns=feature_names)
irisDF['target']=iris.target

dbscan = DBSCAN(eps=0.6, min_samples=8, metric='euclidean')
dbscan_labels = dbscan.fit_predict(iris.data)
irisDF['dbscan_cluster'] = dbscan_labels

iris_result = irisDF.groupby(['target'])['dbscan_cluster'].value_counts()
print(iris_result)
```

* dbsacn_cluster 값에 0, 1 외에도 -1이 군집 레이블로 있음 -> 노이즈 속하는 군집
* 0과 1 두 개의 군집으로 군집화되었음 -> 타겟이 3개 유형인데 두 개 군집이라 해서 효율이 떨어지는 게 아님
* DBSCAN은 군집의 개수를 알고리즘에 따라 자동으로 지정 -> DBCASN에서 군집 개수를 지정하는 것은 무의미
* 붓꽃 데이터셋은 군집을 3개보단 2개로 하는 게 군집화에선 효율적
