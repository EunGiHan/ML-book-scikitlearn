# NMF
* 낮은 팽크를 통한 행렬근사(Low-Rank Approximation) 방식의 변형 (Truncated SVD와 유사하게!)
* 원본 행렬 내의 모든 원소 값이 모두 양수라는 게 보장되면 -> 간단하게 두 개  기반 양수 행렬로 분해 가능
  * 원본행렬 V -> W, H로 근사해 분해 됨

* 행렬 분해(Matrix Factorization)
  * 일반적으로 SVD와 같은 행렬 분해 기법 통칭
  * 길고 가는 행렬 W(원본행렬 행 크기 같고 열크기 보다 작음) & 작고 넓은 행렬 H(원본행렬 열 크기 같고 행 크기 보다 작음)로 분해됨
  * 분해 행렬 W, H은 잠재 요소를 특성으로 가짐
    * W => 원본 행 대해 이 잠재요소 값이 얼마나 되는지
    * H => 잠재요소가 원본 열(원본 속성)로 어떻게 구성되어 있는지
    


# 활용처
* SVD와 유사하게, 차원 축소를 통한 잠재 요소 도출 -> 이미지 변환 및 압축, 텍스트 토픽 및 도출 등에 사용
* 이미지 압축을 통한 패턴 인식
* 텍스트 토픽 모델링 기법
* 문서 유사도 및 클러스터링
* 추천 -> 잠재요소 기반 추천방식

```python
### 모듈 임포트

from sklearn.decomposition import NMF
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
iris_ftrs = iris.data

nmf = NMF(n_components=2)
nmf.fit(iris_ftrs)
iris_nmf = nmf.transform(iris_ftrs)

plt.scatter(x=iris_nmf[:, 0], y=iris_nmf[:, 1], c=iris.target)
plt.xlabel('NMF component 1')
plt.ylabel('NMF component 2')
```
